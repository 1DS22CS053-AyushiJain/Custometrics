```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data = pd.read_csv("marketing_campaign.csv", sep="\t")
```

```python
data.head()
```

```python
data.info()
```

```python
data = data.dropna()
print("The total number of data-points after removing the rows with missing values are:", len(data))
```

```python
print('Rows: {}'.format(data.shape[0]))
print('Columns: {}'.format(data.shape[1]))
```

```python
data.isnull().sum()
```

```python
data["Dt_Customer"] = pd.to_datetime(data["Dt_Customer"], format="%d-%m-%Y")
dates = data["Dt_Customer"].dt.date
print("The newest customer's enrolment date in the records:", dates.max())
print("The oldest customer's enrolment date in the records:", dates.min())
```

```python
days = []
d1 = max(dates) #taking it to be the newest customer
for i in dates:
    delta = d1 - i
    days.append(delta)
data["Customer_For"] = days
data["Customer_For"] = pd.to_numeric(data["Customer_For"], errors="coerce")
```

```python
data.columns
```

```python
print('Total categories in the feature Martial_Status:\n', data['Marital_Status'].value_counts(), '\n')
print('Total categories in the feature education:\n', data['Education'].value_counts())
```

```python
data.describe()
```

```python
data['Age'] = 2014-data['Year_Birth']
data["Spent"] = data["MntWines"]+ data["MntFruits"]+ data["MntMeatProducts"]+ data["MntFishProducts"]+ data["MntSweetProducts"]+ data["MntGoldProds"]
```

```python
data['Living_With'] = data['Marital_Status'].replace({'Married':'Partner', 'Together':'Partner', 'Absurd': 'Alone', 'Widow':'Alone', 'YOLO': 'Alone', 'Divorced':'Alone', 'Single':'Alone'})
```

```python
data=data.rename(columns={"MntWines": "Wines","MntFruits":"Fruits","MntMeatProducts":"Meat","MntFishProducts":"Fish","MntSweetProducts":"Sweets","MntGoldProds":"Gold"})
```

```python
data['Children'] = data['Kidhome']+data['Teenhome']
```

```python
data['Family_Size'] = data['Living_With'].replace({'Alone':1, 'Partner':2})+data['Children']
```

```python
data['Is_Parent'] = np.where(data['Children']>0, 1, 0)
```

```python
data['Education']=data['Education'].replace({'Basic':'UnderGraduate', '2n Cycle':'UnderGraduate', 'Graduation':'Graduate', 'Master':'PostGraduate', 'PhD':'Postgraduate'})
```

```python
to_drop = ["Marital_Status", "Dt_Customer", "Z_CostContact", "Z_Revenue", "Year_Birth", "ID" ]
data = data.drop(to_drop, axis=1)
```

```python
selected_columns = ['Income', 'Kidhome', 'Teenhome', 'Recency', 'Wines',
                    'Fruits', 'Meat', 'Fish', 'Sweets', 'Gold', 'NumDealsPurchases',
                    'NumWebPurchases', 'NumCatalogPurchases', 'NumStorePurchases',
                    'NumWebVisitsMonth', 'AcceptedCmp3', 'AcceptedCmp4', 'AcceptedCmp5',
                    'AcceptedCmp1', 'AcceptedCmp2', 'Complain', 'Response', 'Age', 'Children',
                    'Family_Size', 'Is_Parent', 'Customer_For']
corrmat = data[selected_columns].corr()
plt.figure(figsize=(20,20))
sns.heatmap(corrmat, annot=True, cmap='Spectral', center=0)
```

```python
data.head()
```

```python
product_cols = ["Wines", "Fruits", "Meat", "Fish", "Sweets", "Gold"]
data_filtered = data[product_cols + ["Family_Size"]]
def plot_product_distribution(product):
  plt.figure(figsize=(10, 6))
  sns.boxplot(
      x="Family_Size",
      y=product,
      showmeans=True,
      data=data_filtered
  )
for product in product_cols:
  plot_product_distribution(product)
```

```python
data["Product"] = np.where(data["Wines"] > 0, "Wines", 
                             np.where(data["Fruits"] > 0, "Fruits", 
                                      np.where(data["Meat"] > 0, "Meat", 
                                               np.where(data["Sweets"] > 0, "Sweets", "Other"))))
pivot_table = data.pivot_table(values='Sweets', index='Family_Size', columns='Product')  
plt.figure(figsize=(8, 6))
sns.heatmap(pivot_table, annot=True, cmap="viridis")
plt.title("Spending on Different Products by Family Size")
plt.xlabel("Family Size")
plt.ylabel("Product")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
data['Recency_Cut'] = pd.qcut(data['Recency'], 5)
data['Recency_Cut']
```

```python
R_bins = [-0.001,9.0,19.0,39.0,58.6,79.0,99.0]
R_labels = [0,1,2,3,4,5]
data['Recency_C'] = pd.cut(data['Recency'], bins = R_bins, labels=R_labels)
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(data['Family_Size'], data['Income'], c=data['Recency_C'], cmap='viridis')
plt.title("Relationship Between Family Size, Income, and Recency")
plt.xlabel("Family Size")
plt.ylabel("Income")
plt.colorbar(label='Recency')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
continuous_columns = ['Age', 'Income', 'Spent']
categorical_columns = ['Living_With', 'Children', 'Family_Size', 'Complain', 'Education']
all_columns = continuous_columns + categorical_columns
num_rows = (len(all_columns) + 1) // 2

fig, axes = plt.subplots(nrows=num_rows, ncols=2, figsize=(10, num_rows * 3))
axes = axes.flatten()  # Flatten the axes array to make indexing easier

for i, col in enumerate(all_columns):
    if col in continuous_columns:
        sns.histplot(data=data, x=col, bins=20, ax=axes[i], kde=True, color='blue')
        axes[i].set_title(f'Histogram of {col}')
    elif col in categorical_columns:
        sns.countplot(data=data, x=col, ax=axes[i], color='blue')
        axes[i].set_title(f'Count Plot of {col}')
    axes[i].set_xlabel(col)
    axes[i].set_ylabel('Frequency')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 4))
sns.regplot(x='Income', y='Spent', data=data, scatter_kws={'s': 20}, line_kws={'color': 'red'})

plt.title('Scatter Plot of Income vs. Spent with Trend Line')
plt.xlabel('Income')
plt.ylabel('Spent')
plt.show()
```
